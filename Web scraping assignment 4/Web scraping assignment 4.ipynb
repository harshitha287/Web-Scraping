 Q1. Scrape the details of most viewed videos on YouTube from Wikipedia 

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
time.sleep(3)
```

```python
rank = []
name = []
artist = []
up_date = []
views = []  
```

```python
top_vid = driver.find_element_by_xpath('//li[@class="toclevel-1 tocsection-2"]//a')
top_vid.click()
time.sleep(1)

#using for loop to loop over all the rows at once
for i in range (1, 31):                                              
    xp = '//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr['
    rank_path = driver.find_element_by_xpath(xp + str(i) + ']//td[1]')
    name_path = driver.find_element_by_xpath(xp + str(i) + ']//td[2]')
    artist_path = driver.find_element_by_xpath(xp + str(i) + ']//td[3]')
    up_date_path = driver.find_element_by_xpath(xp + str(i) + ']//td[4]')
    views_path = driver.find_element_by_xpath(xp + str(i) + ']//td[5]')
    
    rank.append(rank_path.text)
    name.append(name_path.text)
    artist.append(artist_path.text)
    up_date.append(up_date_path.text)
    views.append(views_path.text)
```

```python
wiki = pd.DataFrame({})                                            
wiki['Rank'] = rank
wiki['Name'] = name
wiki['Artist'] = artist
wiki['Upload Date'] = up_date
wiki['Views'] = views
wiki.set_index('Rank', inplace = True)
wiki['Name'] = wiki['Name'].apply(lambda x: x[1:-5])        #removing quotation marks and [i]
wiki
```

```python
eco = driver.find_element_by_xpath('//div[@class="dropdown"][2]')
hover = ActionChains(driver).move_to_element(eco)
hover.perform()                             
#this will hover the mouse cursor over the element so drop down menu will appear

ind = driver.find_element_by_xpath('//div[@class="dropdown"][2]//a[3]')
ind.click()
time.sleep(2)                               
#now we are able to access dropdown menu options

gdp_page = driver.find_element_by_xpath('//ul[@style="list-style-type:none;margin-left:20px;"]//li[1]')
gdp_page.click()
time.sleep(2)
```

```python
name=[]
designation=[]
company=[]
skill=[]
location=[]
```

```python
name=[]
try:
    names=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[1]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:
    name.append('No details available')
print(len(name))

designation=[]
try:
    designation_list=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/span[1]")
    for i in designation_list:
        designation.append(i.text)
except NoSuchElementException:
    designation.append('No details available')
print(len(designation))

company=[]
try:
    companies=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[2]")
    for i in companies:
        company.append(i.text)
except NoSuchElementException:
    company.append('No details available')
print(len(company))

skill=[]
try:
    skills=driver.find_elements_by_xpath("//div[@class='recInfo']/div[2]")
    for i in skills:
        skill.append(i.text)
except NoSuchElementException:
    skill.append('No details available')
print(len(skill))
```

Q10. Details of Datasets from UCI machine learning repositories

```python
driver = webdriver.Chrome(r"C:\Users\Lenovo\Downloads\chromedriver_win32\chromedriver.exe")
url = 'https://archive.ics.uci.edu/'
driver.get(url)
time.sleep(3)
```

```python
ds = driver.find_element_by_xpath('//font[@size="3"]')
ds.click()
time.sleep(2)
```

```python
ds_name = []
dtype = []
task = []
att = []
no_inst = []
no_att = []
year = []
```

```python
total_tag = driver.find_element_by_xpath('//table[@width="100%"]//p[@class="big"]//b')
total = int (total_tag.text) + 2

for i in range (2, total):          
#creating a list of list in which each element will contain all the column info in a list
    
    data_tag = driver.find_elements_by_xpath('//table[@border="1"]//tr[' + str(i) + ']//p')
    ds_name.append(data_tag[0].text)
    dtype.append(data_tag[1].text)
    task.append(data_tag[2].text)
    att.append(data_tag[3].text)
    no_inst.append(data_tag[4].text)
    no_att.append(data_tag[5].text)
    year.append(data_tag[6].text)
```

```python
uci = pd.DataFrame({})                                            
uci['Dataset name'] = ds_name
uci['Data type'] = dtype
uci['Task'] = task
uci['Attribute type'] = att
uci['No of instances'] = no_inst
uci['No of attribute'] = no_att
uci['Year'] = year
uci
```
